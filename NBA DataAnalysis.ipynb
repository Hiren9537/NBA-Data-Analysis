```python
data1 = data.drop(['position'],axis = 'columns')
data1.head()
```

```python
data1.isnull().sum()
```

# Dropping Null Values

```python
data2 = data1.dropna()
data2.isnull().sum()
```

```python
data2.shape
```

```python
data2
```

# Replacing ( - ) to ( . ) in height 

```python
data2.height = data2.height.apply(lambda x: x.replace('-','.'))
data2.head()
```

```python
data2.groupby('college')['college'].agg('count')
```

```python
data2.groupby('year_start')['year_start'].agg('count')
```

```python
startedbetween1950 = data2[data2['year_start'] <= 1950]['year_start']
startedbetween1950
```

```python
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (20,10)
plt.hist(startedbetween1950,rwidth = 0.5)
plt.xlabel('Students started playing basketball(In Year)', size = 12,alpha = 1)
plt.ylabel('Count',size = 12)
```
